```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
dfsh = pd.read_csv("Algerian_forest_fires_cleaned.csv",)
dfsh.head()
```

Data Preparation

```python
dfsh.info()
```

```python
dfsh.describe()
```

```python
dfsh.duplicated().sum()
```

```python
dfsh.plot(
    kind='box',
    subplots=True,
    sharey=False,
    figsize=(20, 7)
)
plt.subplots_adjust(wspace=1)
plt.show()
```

```python
dfsh.columns
```

```python
#Removing outliers for every other columns other than RH
for column in ['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Region']:
    Q1 = dfsh[column].quantile(0.25)
    Q3 = dfsh[column].quantile(0.75)
    IQR = Q3 - Q1
    lowerBound = Q1 - 1.5 * IQR
    upperBound = Q3 + 1.5 * IQR
    dfsh = dfsh[(dfsh[column] >= lowerBound) & (dfsh[column] <= upperBound)]
```

```python
dfsh.shape
```

```python
dfsh.plot(
    kind='box',
    subplots=True,
    sharey=False,
    figsize=(20, 7)
)
plt.subplots_adjust(wspace=1)
plt.show()
```

```python
corr_matrix = dfsh.corr()
plt.figure(figsize=(12, 7))
sns.heatmap(corr_matrix, annot=True,  cmap='coolwarm')
plt.show()
```

```python
#Taking all columns for training the first model
X = dfsh[['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'Region']]
Y = dfsh['FWI']
```

```python
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
sc = StandardScaler()
trainXsc = sc.fit_transform(trainX)
testXsc = sc.transform(testX)
```

Model Training 1

```python
#Linear Regression
lrModel1 = LinearRegression()
```

```python
lrModel1.fit(trainXsc, trainY)
```

```python
lrModel1.coef_, lrModel1.intercept_
```

```python
yPred = lrModel1.predict(testXsc)
```

```python
#Metrics for Model 1
mse = mean_squared_error(testY,yPred)
mae = mean_absolute_error(testY,yPred)
rsq = r2_score(testY,yPred)
print("LINEAR REGRESSION Model 1 MSE : ", mse)
print("LINEAR REGRESSION Model 1 MAE : ", mae)
print("LINEAR REGRESSION Model 1 R2 : ", rsq)
```

Model Training 2

```python
#After considering correlation and Coefficients from model 1 removed some columns
X = dfsh[['month', 'Temperature', 'RH', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI']]
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.2, random_state=42)
sc = StandardScaler()
trainXsc = sc.fit_transform(trainX)
testXsc = sc.transform(testX)
```

```python
lrModel2 = LinearRegression()
lrModel2.fit(trainXsc, trainY)
lrModel2.coef_, lrModel2.intercept_
```

```python
yPred = lrModel2.predict(testXsc)
mse = mean_squared_error(testY,yPred)
mae = mean_absolute_error(testY,yPred)
rsq = r2_score(testY,yPred)
print("LINEAR REGRESSION Model 2 MSE : ", mse)
print("LINEAR REGRESSION Model 2 MAE : ", mae)
print("LINEAR REGRESSION Model 2 R2 : ", rsq)
```

Pickling The Model

```python
import pickle
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(lrModel2, file)
with open('scaler_model.pkl', 'wb') as file:
    pickle.dump(sc, file)
```

Reading Pickled Files

```python
with open('linear_regression_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

with open('scaler_model.pkl', 'rb') as file2:
    loaded_scaler = pickle.load(file2)
```

Predicting On Unseen Data

```python
dataX = [[6,31,64,86.8,17.8,71.8,6.7,21.6],
[8,34,53,89,21.6,80.3,9.2,25.8],
[9,32,55,89.1,25.5,88.5,7.6,29.7],
[6,32,47,79.9,18.4,84.4,2.2,23.8],
[6,33,50,88.7,22.9,92.8,7.2,28.3],
[9,29,68,59.9,2.5,8.6,1.1,2.9],
[9,27,75,55.7,2.4,8.3,0.8,2.8],
[7,32,76,63.1,2.6,9.2,1.3,3],
]

dataY = [10.6,15,13.9,3.9,12.9,0.4,0.3,0.5]
```

```python
scaledX = loaded_scaler.transform(dataX)
```

```python
yPred = loaded_model.predict(scaledX)
mse = mean_squared_error(dataY,yPred)
mae = mean_absolute_error(dataY,yPred)
rsq = r2_score(dataY,yPred)
print("Linear Model MSE : ", mse)
print("Linear Model MAE : ", mae)
print("Linear Model R2 : ", rsq)
```

Conclusion

An R-squared score of 0.99, which indicates that 99% of the variation of the dependent variable can be predicted, was obtained after training Linear Regression Model 1. The average difference from the real values was indicated by the Mean Absolute Error (MAE) of 0.38 and the Mean Squared Error (MSE) of 0.23, which indicate the error from the actual values.
Then, using the knowledge from Linear Model 1 to eliminate some columns, we trained Linear Regression Model 2, which also produced an R-squared score of 0.99. This implies that Model 2 can predict 99 percent of the variation of the dependent variable. In contrast to Model 1, the MSE and MAE were marginally lower at 0.22 and 0.33, respectively, suggesting a minor performance improvement.
To improve our models, we can experiment with alternative combinations of independent variables that have strong correlation and coefficient values. Improvements in model performance and forecast accuracy may result from this iterative process.
